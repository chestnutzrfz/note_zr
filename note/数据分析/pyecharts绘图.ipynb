```python
"""柱状图"""

# 导包
import pandas as pd
from pyecharts.charts import Bar

# 创建柱状图的绘制对象
bar = Bar()

# 定义x,y轴的数据
bar.add_xaxis(['数学', '英语', '语文'])
bar.add_yaxis('科目', [88, 90, 40])
bar.add_yaxis('小红', [88, 90, 40])

# 数据渲染到html页面
bar.render_notebook()
```

```python
"""柱状图2"""

# 导包
import pandas as pd
from pyecharts.charts import Bar
import pyecharts.options as opts

# 创建柱状图的绘制对象
bar = (
    # Bar(init_opts=opts.InitOpts(theme=ThemeType.WESTEROS))
    Bar()
    .add_xaxis(['数学','英语','语文'])
    .add_yaxis('科目',[90,32,60])
    .set_global_opts(title_opts=opts.TitleOpts(title='主标题',subtitle='副标题'))
)

# 数据渲染到html页面
bar.render_notebook()
```

```python
"""柱状叠图"""

# 导包
import pandas as pd
from pyecharts.charts import Bar
import pyecharts.options as opts

# 创建柱状图的绘制对象
bar = Bar()

# 定义x,y轴的数据
bar.add_xaxis(['数学', '英语', '语文'])
bar.add_yaxis('科目', [88, 90, 40])
bar.add_yaxis('小红', [88, 90, 40])

# 设置柱状堆叠
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))

# 将制作的图形结果html保存到本地
bar.render('柱状堆叠.html')
```

```python
"""饼图"""

# 导包
from pyecharts.charts import Pie

# 创建饼图的绘制对象
pie = Pie()

# 数据
x = ['数学','语文','英语','物理','化学']
y = [43,54,76,87,97]

# 制作饼图的数据需要转换成元组在列表中
data = list(zip(x,y))

# 将数据传入到饼图对象中
pie.add('科目',data)

# 数据渲染
pie.render_notebook()
```

```python
"""玫瑰图"""

# 导包
from pyecharts.charts import Pie

# 创建饼图的绘制对象
pie = Pie()

# 数据
x = ['数学','语文','英语','物理','化学']
y = [43,54,76,87,97]

# 制作玫瑰图的数据需要转换成元组在列表中
data = list(zip(x,y))

# 将数据传入到饼图对象中
pie.add('科目',data,rosetype='1')
# resetype='1'  制作玫瑰图

# 数据渲染
pie.render_notebook()
```

```python
"""制作仪表盘"""

# 导包
from pyecharts.charts import Gauge

# 创建仪表盘对象
gg = Gauge()

# 数据
x = ['数学','语文','英语','物理','化学']
y = [43,54,76,87,97]

# 制作仪表盘的数据需要转换成元组在列表中
data = list(zip(x,y))

# 将数据传入到仪表盘对象中
gg.add('科目',data)

# 数据渲染
gg.render_notebook()
```

```python
"""折线图"""

# 导包
from pyecharts.charts import Line
import pyecharts.options as opts

# 创建折线图对象
line = Line()

# 数据
x = ['数学','语文','英语','物理','化学']
y = [43,54,76,87,97]
yy = [2*i for i in y]

# 定义x轴
# 注意: 此处的参数需要指定传参
line.add_xaxis(xaxis_data=x)
line.add_yaxis(series_name='小明',y_axis=y)
line.add_yaxis(series_name='小红',y_axis=yy)

# 添加标题
line.set_global_opts(title_opts=opts.TitleOpts(title='主标题',subtitle='副标题'))

# 数据渲染
line.render_notebook()
```

```python
"""词云图"""

# 导包
from pyecharts.charts import WordCloud

# 创建对象
w = WordCloud()

# 数据
x = ['数学','语文','英语','物理','化学']
y = [43,54,76,87,97]

# 制作词云图的数据需要转换成元组在列表中
data = list(zip(x,y))

# 将数据传入到词云图对象中
w.add('科目',data)

# 数据渲染
w.render_notebook()
```

```python
"""散点图"""

# 导包
from pyecharts.charts import Scatter

# 创建对象
s = Scatter()

# 将数据传入到散点图对象中
s.add_xaxis(['数学','语文','英语','物理','化学'])
s.add_yaxis('小明',[43,54,76,87,97])
s.add_yaxis('小红',[65,94,43,65,86])

# 数据渲染
s.render_notebook()
```
