```python
"""折线图"""

# 导包
import matplotlib.pyplot as plt

# 创建画布
# fig = plt.figure(figsize=(6,10),facecolor='b',dpi=500)
fig = plt.figure(figsize=(6,10),dpi=500)
# figsize  调整画布尺寸比例大小
# facecolor  指定画布的北京颜色,一般是以代码的英文首字母
# dpi  画布的分辨率 原则上来说, 值越大图形越清晰

# 解决中文显示的问题
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

# 指定图形的标题
plt.title('爬虫')




# 数据
x = [1,2,3,4,5,6,7,8,9]
y = [6,5,3,2,4,5,6,7,8]
# 若要应用于其他项目, 修改x,y这两个列表的数据即可
yy = [2*i for i in y]

# 绘制图像
plt.plot(x,y, linestyle='-.',color='r')
# 绘制多条折线
plt.plot(x,yy,linestyle='--',color='b')
# linestyle=='--'   指定折线为虚线
# color  指定折线的颜色

# 显示点的数据值
for x1,y1 in zip(x,y):
    plt.text(x1,y1,f'{(x1,y1)}',fontdict={'fontsize':14})
for x1,y2 in zip(x,yy):
    plt.text(x1,y2,f'{(x1,y2)}',fontdict={'fontsize':14})

# 添加x,y轴的说明(标注)
plt.xlabel('x')
plt.ylabel('y')

# 保存
plt.savefig('折线.png')

# 显示
plt.show()
```

```python
"""柱状图"""

# 导包
import matplotlib.pyplot as plt

# 创建画布
# fig = plt.figure(figsize=(6,10),facecolor='b',dpi=500)
fig = plt.figure(figsize=(6,10),dpi=500)
# figsize  调整画布尺寸比例大小
# facecolor  指定画布的北京颜色,一般是以代码的英文首字母
# dpi  画布的分辨率 原则上来说, 值越大图形越清晰

# 解决中文显示的问题
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

# 指定图形的标题
plt.title('爬虫')



# x轴：月份列表
x = [i for i in range(1, 13)]
# y轴：
y = [5, 4, 2, 7, 8, 2, 4, 9, 2, 4, 5, 5]
yy = [i*2 for i in y]

# 绘制柱状图
plt.bar(x, y, width=0.4, color='#000000', tick_label=[f"{i}月"for i in x], label='爬')
plt.bar([0.2+i for i in x], yy, width=0.2, color='r', label='虫', hatch='\\')
# width: 指定柱状的宽度
# color: 指定柱状的颜色, 列表是一对一指定颜色，字符串是指定所有柱状的颜色
# tick_label：指定x轴的文字
# hatch='\\'给柱状添加添加特殊绘制

# 改变x轴文字显示方向
plt.xticks(rotation=270)
# 颜色码地址：http://wjhsh.net/weiyinfu-p-9164345.html

for x1, y2 in zip(x, y):
    plt.text(x1, y2, f'{y2}元',  fontdict={'fontsize': 14})
for x1, y3 in zip(x, yy):
    plt.text(x1, y3, f'{y3}元',  fontdict={'fontsize': 14})
    
# 添加标题
plt.title('爬虫')

# 添加说明
plt.xlabel('月份')
plt.ylabel('价格')

# 添加右上角柱状声明
plt.legend()

# 保存图形
plt.savefig('柱状.png')

# 显示图形
plt.show()
```

```python
"""饼图"""

# 导包
import matplotlib.pyplot as plt

# 创建画布
# fig = plt.figure(figsize=(6,10),facecolor='b',dpi=500)
fig = plt.figure(figsize=(6,10),dpi=500)
# figsize  调整画布尺寸比例大小
# facecolor  指定画布的北京颜色,一般是以代码的英文首字母
# dpi  画布的分辨率 原则上来说, 值越大图形越清晰

# 解决中文显示的问题
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

# 指定图形的标题
plt.title('爬虫')



# 饼图：一般是用来观察百分占比的

x = ['语文', '数学', '英语']
y = [88, 98, 30]

# 绘制饼图
plt.pie(y,labels=x,autopct='%1.1f%%')
# 第一个参数用于计算饼图扇形分布
# labels  显示所在扇区的描述
# autopct  显示百分占比

# 将图形保存到本地
plt.savefig('饼图.png')

plt.show()
```
